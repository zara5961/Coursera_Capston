```python
# import pandas library
import pandas as pd
import numpy as np
```

```python
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
# Read the online file by the URL provides above, and assign it to variable "df"
df = pd.read_csv(URL, header=None)
```

```python
# Show the first 5 rows using dataframe.head() method just to check 
print("The first 5 rows of the dataframe") 
df.head(5)
```

```python
# Show the last 5 rows using dataframe.head() method just to check 
print("The Last 5 rows of the dataframe") 
df.tail(5)
```

```python
# To better describe our data we can introduce a header, this information is available at: "https://archive.ics.uci.edu/ml/datasets/Automobile"

# Create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
# We replace headers and recheck our data frame
df.columns = headers
df.head(5)
```

```python
# we need to replace the "?" symbol with NaN 
df.replace("?" , np.NAN, inplace= True)
df.head(10)
```

```python
# Evaluating for Missing Data: .isnull() or .notnull() method; The output is a boolean value 
missing_data = df.isnull()
missing_data.head(5)
```

```python
# Count missing values using a for loop; "True" represents a missing value

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
# For "normalized-losses", "stroke", "bore", "horsepower" and "peak-rpm"  missing data, replace them with mean

# Calculate the average of the column "normalized-losses"
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)


# Calculate the mean vaule for "stroke" column
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)


# Calculate the mean value for "bore" column
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)


# Calculate the mean value for the "horsepower" column
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)


# Calculate the mean value for the "peak-rpm" column
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
# Replace "NaN" by mean value in "normalized-losses" column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)



# Replace "NaN" by mean value in "stroke" column
df["stroke"].replace(np.nan, avg_stroke, inplace = True)



# Replace "NaN" by mean value in "bore" column
df["bore"].replace(np.nan, avg_bore, inplace=True)



# Replace "NaN" by mean value in "horsepower" column
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)



# Replace "NaN" by mean value in "peak-rpm" column
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)


df.head(10)
```

```python
# For "num-of-doors" missing data, replace with most frequent data

# To see which values are present in a particular column
df['num-of-doors'].value_counts()
```

```python
# Or Use the ".idxmax()" to calculate the most common type automatically:
df['num-of-doors'].value_counts().idxmax()
```

```python
# Replace the missing "num-of-doors" values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# FOR "price" missing data, we'll drop all rows that do not have data (price is a main part of the problem)

# Drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)
df.head(10)
```

```python
# Reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
df.head(10)
```

```python
# Last step for cleaning is checking all data type
df.dtypes
```

```python
# Convert data types to proper format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
# Standardization: "city-mpg" and "highway-mpg" columns which are represented by mpg (miles per gallon) to L/100km 
df['city-L/100km'] = 235/df["city-mpg"]
df["highway-L/100km"] = 235/df["highway-mpg"]
df.head()
```

```python
# Normalization: scale the columns "length", "width" and "height" in a ranges from 0 to 1.
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()

# show the scaled columns
df[["length","width","height"]].head()
```

```python
# Convert data to correct format 

df["horsepower"]=df["horsepower"].astype(int, copy=True)
df["horsepower"]
```

```python
# Binning: "horsepower" is a valued variable but our pric e is only related to high, medium, and little horsepower 
# 3 bins of equal size bandwidth 
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']

# We apply the function "cut" to determine values to each bins
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(10)
```

```python
# Number of vehicles in each bin
df["horsepower-binned"].value_counts()
```

```python
#Plot the data

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

# Plot Barchart 
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Plot Histogram
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# One hot encoding
# Indicator variable: "fuel-type" has two unique values, "gas" or "diesel", change to Indicator value and change names
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# Merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# Drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```

```python
# Indicator variable: "aspiration" has two unique values, "std" or "turbo", change to Indicator value and change names
dummy_variable_2 = pd.get_dummies(df['aspiration'])
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
dummy_variable_2.head()
```

```python
# Merge the new dataframe to the original datafram
df = pd.concat([df, dummy_variable_2], axis=1)

# Drop original column "aspiration" from "df"
df.drop('aspiration', axis = 1, inplace=True)
df.head()
```
